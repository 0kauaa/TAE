# Dados

```r
data <- read.csv('https://raw.githubusercontent.com/cibelerusso/Datasets/refs/heads/main/DadoseDecisoes.csv')
tail(data)
```

```r
encoded <- fastDummies::dummy_cols(data, select_columns = c('Gênero', 'Home_Office'),
                                       remove_first_dummy = TRUE,
                                       remove_selected_columns = TRUE)
```

```r
attach(encoded)
```

```r
x <- subset(encoded, select = -c(Departamento))
y <- as.numeric(factor(Departamento))
```

# KNN

```r
# linha teste (aleatoria)
n <- sample(1:nrow(x), 1)
linha <- x[n, ]
```

```r
# modelo
knn_model <- knn(
  train = x[-n, ],
  test = linha,
  cl = y[-n],
  k = 20
)
```

```r
print("Célula testada (dados originais):")
print(data[n,])

departamentos <- levels(factor(data$Departamento))
labelout <- as.numeric(as.character(knn_model))
departamento_predito <- departamentos[labelout]

print("Departamento predito:")
print(departamento_predito)

print("Departamento real:")
print(data[n,]$Departamento)
```
